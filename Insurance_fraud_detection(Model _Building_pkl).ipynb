# Application of final model for Pickle file creation

```python
#import necessary libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
fraud_ml= pd.read_csv('C:/Users/HARESH/Downloads/Final project/Fraud_data_final_new.csv')
```

```python
fraud_ml.shape
```

```python
fraud_ml.head()
```

```python
fraud_ml=fraud_ml.rename({'Home or self care,':'Home_care'},axis=1)
fraud_ml=fraud_ml.rename({'Mortality risk':'Mortality_risk'},axis=1)
fraud_ml=fraud_ml.rename({'Emergency dept_yes/No':'Emergency_dept_yes_No'},axis=1)
```

```python
fraud_ml=fraud_ml.drop(['Unnamed: 0'],axis=1)
fraud_ml
```

```python
x_res = fraud_ml.drop(['Result'], axis=1)
y_res = fraud_ml['Result']
```

```python
from sklearn.ensemble import RandomForestClassifier as RF
model = RF(n_jobs=2, random_state=40)
model.fit(x_res,y_res)
```

```python
##  create a pickle file using serialization
import pickle
pickle_out=open("model_pkl.pkl","wb")
pickle.dump(model,pickle_out)
pickle_out.close()
```
